```python
income = pd.read_csv("./Data/updates_irs.csv")
```

```python
income.drop(columns="Unnamed: 0", inplace = True)
```

```python
income.rename(columns={"zipcode": "Zip", "STATE": "State"}, inplace = True)
```

```python
#we only need zipcode ,avg_gross_income and avg_total_income
cols_to_keep = ["Zip", "avg_gross_income", "avg_total_income"]
income = income[cols_to_keep]
```

```python
income
```

```python
#now let's merge our dataframes
df = moto.join(income.set_index("Zip"), on = "Zip")
```
